```python
import os
from omegaconf import OmegaConf

import pandas as pd

from sklearn.metrics import accuracy_score
```

```python
NUM_CLASSES = 5
```

```python
config = OmegaConf.load("/Users/sem-k32/10_sem/n_ode/experiment/ecg5000/config.yaml")
```

```python
cls_result = [
    pd.read_csv(
        os.path.join(config.results_dir, f"cls_{target_label}_{pred_label}.csv")
    )
    for target_label in range(NUM_CLASSES)
    for pred_label in range(NUM_CLASSES)
]
cls_result = pd.concat(cls_result, ignore_index=True)
cls_result.head()
```

```python
pred_argmin = cls_result.groupby("target", "traj_num")["distance"].idxmin()
pred = cls_result[pred_argmin]["pred"]
true = cls_result[pred_argmin]["target"]
accuracy_score(true, pred)
```
