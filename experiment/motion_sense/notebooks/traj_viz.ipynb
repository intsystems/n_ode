```python
from omegaconf import OmegaConf
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import scipy.linalg as linalg
from sklearn.decomposition import PCA

from experiment.raw_data_loading import create_time_series, set_data_types
```

```python
SEED = 123
np.random.seed(SEED)
```

```python
config = OmegaConf.load("config.yaml")
```

```python
subjects = pd.read_csv(config.data_dir + "/data_subjects_info.csv")
subjects.head()
```

```python
# subjects_sample = subjects["code"].sample(2).tolist()
# subjects_sample
subjects_sample = [0, 4]
subjects_sample
```

```python
for act, activity_codes in config.activity_codes.items():
    print(act)
    act_code = activity_codes[0]
    for data_type in ["attitude", "rotationRate", "userAcceleration"]:
        print(data_type)
        dt_columns = set_data_types([data_type])
        act_labels = [act]
        trial_codes = [[act_code]]
        ts_df = create_time_series(
            config.data_dir, dt_columns, act_labels, trial_codes, mode="raw"
        )
        ts_df = ts_df[ts_df["id"].isin(subjects_sample)]
        ts_df = ts_df.iloc[::20]
        ts_df["t"] = ts_df.groupby("id", group_keys=False).apply(
            lambda d: pd.Series(range(d.shape[0]), index=d.index)
        )
        # print(ts_df.head())
        # break
        for ts_name in ts_df.columns[:3]:
            px.line(
                ts_df, x="t", y=ts_name, color="id"
            ).show()
```
