```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
import pandas as pd
import scipy.linalg as linalg
from sklearn.decomposition import PCA
```

```python
traj_data = pd.read_csv("ups.csv")[["lineKey", "lineVal", "step"]]
```

```python
traj_data.head()
```

```python
mpl.rcParams['figure.dpi'] = 300
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.markersize'] = 12
plt.rcParams['xtick.labelsize'] = 24
plt.rcParams['ytick.labelsize'] = 24
plt.rcParams['legend.fontsize'] = 24
plt.rcParams['axes.titlesize'] = 36
plt.rcParams['axes.labelsize'] = 24

fig, ax = plt.subplots(figsize=(15, 8))

ax.plot(
    traj_data[traj_data["lineKey"] == "true"]["step"].values, 
    traj_data[traj_data["lineKey"] == "true"]["lineVal"].values,
    label="true",
    color="#3a0ca3",
    linestyle="--"
)

ax.plot(
    traj_data[traj_data["lineKey"] == "pred"]["step"].values, 
    traj_data[traj_data["lineKey"] == "pred"]["lineVal"].values,
    label="pred",
    color="#4361ee"
)

ax.grid(True)
ax.legend()
ax.set_xlabel("$z_1$")
ax.set_ylabel("$z_2$")
ax.set_title("Ups trajectory approximation")
```
